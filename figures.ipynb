```python

# distribution of number of lines
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# inputs: ids
# outputs: heatmap of detection and temporal distribution
split = "a03"
ids_ctrl = np.sort([id for id in data.ids[split] if "ctrl" in id])
ids_virus = np.sort([id for id in data.ids[split] if "virus" in id])
# collect all detection
cir_ctrl = []
ls_det = []
for f in data.ids_to_labels(ids_virus):
    with open(f) as f:
        ls_det += [l.strip() for l in f.readlines()] 
for det in ls_det:
    # split by space
    dets = det.split(" ")
    x = float(dets[1])
    y = float(dets[2])
    w = float(dets[3])
    h = float(dets[4])
    r = (w * h) ** 0.5 / 2 # radius
    cir_ctrl += [(x, y, r)]

from scipy.ndimage import gaussian_filter

# Image size
image_size = (1000, 1000)


# Create an empty image grid
heatmap = np.zeros(image_size)

# Resize the coordinates and plot circles
for x, y, r in cir_ctrl:
    x_resized = int(x * image_size[0])
    y_resized = int(y * image_size[1])
    radius_resized = int(r * image_size[0])

    # Create a grid of the same size as the image
    Y, X = np.ogrid[:image_size[0], :image_size[1]]
    distance = np.sqrt((X - x_resized) ** 2 + (Y - y_resized) ** 2)

    # Add intensity to the heatmap
    heatmap += np.exp(-(distance**2) / (2 * (radius_resized**2)))

# Apply Gaussian filter to smooth the heatmap
# heatmap = gaussian_filter(heatmap, sigma=5)

# Plot the heatmap
plt.imshow(heatmap, cmap='hot', extent=(0, 1, 0, 1))
plt.colorbar(label='Intensity')
plt.title('Heatmap of Circle Distribution')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.axis("off")
plt.show()
```
