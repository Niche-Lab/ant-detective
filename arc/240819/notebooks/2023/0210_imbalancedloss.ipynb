# Test Imbalanced sampler

```python
import torch.nn as nn
import torch
```

```python
loss = nn.HuberLoss(delta=10)
for x in range(20):
    targt = torch.tensor([1.0] * 19 + [float(x)])
    logit = torch.tensor([1.0] * 20)
    print(logit)
    print(targt)
    output = loss(logit, targt)
    print(output)
```
